```python
import pandas as pd
import itertools
import json
```

```python
t_df = pd.read_csv("/Users/samailguliyev/Downloads/DMC-2021-Task/transactions.csv",sep='|')
```

```python
d = {}
l = []

for i,r in t_df[t_df.duplicated(subset=['sessionID'],keep=False)].iterrows():
#     print(r['sessionID'])
    session_id = r['sessionID']
#     print(i)
    if session_id in d:
        d[session_id][r['itemID']] = [r['click'],r['basket'],r['order']]
    else:
        d[session_id] = {r['itemID']:[r['click'],r['basket'],r['order']]}


  

        
d

### Outputs dictionary. Keys - transaction ids /// values - list of items   
```

```python
k = {}
for i in d:
    c = list(itertools.permutations(d[i], 2))
    for j in c:
        if j not in k:
            k[j] = {'co-occured' : 1,
                    'co-clicked' : 0,
                    'co-basketed': 0,
                    'co-bought': 0}
            if (d[i][j[0]][0] >0) & (d[i][j[1]][0] >0):
                k[j]['co-clicked'] = k[j]['co-clicked'] + 1
            if (d[i][j[0]][1] >0) & (d[i][j[1]][1] >0):
                k[j]['co-basketed'] = k[j]['co-basketed'] + 1
            if (d[i][j[0]][2] >0) & (d[i][j[1]][2] >0):
                k[j]['co-bought'] = k[j]['co-bought'] + 1
        else:
            k[j]['co-occured'] = k[j]['co-occured'] + 1
            if (d[i][j[0]][0] >0) & (d[i][j[1]][0] >0):
                k[j]['co-clicked'] = k[j]['co-clicked'] + 1
            if (d[i][j[0]][1] >0) & (d[i][j[1]][1] >0):
                k[j]['co-basketed'] = k[j]['co-basketed'] + 1
            if (d[i][j[0]][2] >0) & (d[i][j[1]][2] >0):
                k[j]['co-bought'] = k[j]['co-bought'] + 1

# Outputs dictionary/// keys - tuple of item pairs (item1,item2)/// values - dictionary of co-occurance 
```
